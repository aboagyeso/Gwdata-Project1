```python
# Dependencies
import tweepy
import json
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
from config import consumer_key, consumer_secret, access_token, access_token_secret

# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python

# using drug overdose deaths as the search term
search_term ="Drug overdose deaths "

# Search for all tweets
public_tweets = api.search(search_term)
public_tweets
```

```python
# View Search Object
# print(public_tweets)

# Loop through all tweets
for tweet in public_tweets["statuses"]:

    # Utilize JSON dumps to generate a pretty-printed json
    print(json.dumps(tweet, sort_keys=True, indent=4, separators=(',', ': ')))
    
```

```python
overdose_tweets = print(tweet["text"])
overdose_tweets    
```

```python
# Run Vader Analysis on Passage
compound = analyzer.polarity_scores('overdose_tweets')["compound"]
pos = analyzer.polarity_scores('overdose_tweets')["pos"]
neu = analyzer.polarity_scores('overdose_tweets')["neu"]
neg = analyzer.polarity_scores('overdose_tweets')["neg"]
```

```python
# Print Analysis
print(f"Compound Score: {'compound'}")
print(f"Positive Score: {'pos'}")
print(f"Neutral Score: {'neu'}")
print(f"Negative Score: {'neg'}")
```
