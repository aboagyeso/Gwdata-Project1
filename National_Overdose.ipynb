```python
#import dependencies 
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
National_Overdose = pd.read_csv("OVERDOSE2.csv")
```

```python
National_Overdose
```

```python
#Summary statistics of overdose data
National_Overdose.describe()
```

```python
#Total death per category
National_Overdose.sum()
```

```python
#National Overdose Trend 
plt.figure(figsize=(18,10))
plt.plot(National_Overdose['YEAR'], National_Overdose['TOTAL'],color='red',label="Male Overdose")
plt.title("TOTAL OVERDOSE DEATHS", fontsize=26)
plt.xlabel("YEAR", fontsize=20)
plt.ylabel("OVERDOSE DEATHS", fontsize=20)
legend = plt.legend(numpoints= 1.5, 
                    markerscale= 1.5, 
                    fontsize= 18, 
                    frameon=True, 
                    edgecolor="black")
plt.show()
```

```python
#Male and Female overdose trend
plt.figure(figsize=(18,10))
plt.plot(National_Overdose['YEAR'], National_Overdose['TOTAL'],color='red',label="Female Overdose")
plt.plot(National_Overdose['YEAR'], National_Overdose['MALE'],color='green',label="Female Overdose")
plt.plot(National_Overdose['YEAR'], National_Overdose['FEMALE'],color='purple',label="Male Overdose")
plt.title("MALE VS FEMALE OVERDOSE DEATHS", fontsize=26)
plt.xlabel("YEAR", fontsize=20)
plt.ylabel("OVERDOSE DEATHS", fontsize=20)
legend = plt.legend(numpoints= 1.5, 
                    markerscale= 1.5, 
                    fontsize= 18, 
                    frameon=True, 
                    edgecolor="black")
plt.show()
```

```python
#OVERDOSE DEATHS BY RACE 
plt.figure(figsize=(18,10))
plt.plot(National_Overdose['YEAR'], National_Overdose['TOTAL'],color='red',label="Female Overdose")
plt.plot(National_Overdose['YEAR'], National_Overdose['WHITE'],color='aqua',label="WHITE OVERDOSE")
plt.plot(National_Overdose['YEAR'], National_Overdose['BLACK'],color='black',label="Male Overdose")
plt.plot(National_Overdose['YEAR'], National_Overdose['HISPANIC'],color='brown',label="Male Overdose")
plt.title("OVERDOSE DEATHS BY RACE", fontsize=26)
plt.xlabel("YEAR", fontsize=20)
plt.ylabel("OVERDOSE DEATHS", fontsize=20)
#legend = plt.legend(numpoints= 1.5, 
                    #markerscale= 1.5, 
                    #fontsize= 18, 
                    #frameon=True, 
                    #edgecolor="black")
plt.show()
```

```python

#OVERDOSE DEATHS BY AGE
plt.figure(figsize=(18,10))
plt.plot(National_Overdose['YEAR'], National_Overdose['TOTAL'],color='red',label="Female Overdose")
plt.plot(National_Overdose['YEAR'], National_Overdose['25-34'],color='blue')
plt.plot(National_Overdose['YEAR'], National_Overdose['45-54'],color='black')
plt.title("OVERDOSE DEATHS BY AGE", fontsize=26)
plt.xlabel("YEAR", fontsize=20)
plt.ylabel("OVERDOSE DEATHS", fontsize=20)
plt.show()
```
