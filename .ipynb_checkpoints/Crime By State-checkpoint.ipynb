```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numpy import array
import matplotlib.dates as mdates
```

```python
# Read CSV from U.S Dept of Justice
data_file = "./Resources/CrimeByState.csv"
crime_data_df = pd.read_csv(data_file)
crime_data_df
```

```python
# Get average violent crime rate from each state from 1999-2014
crime_rate_avg = crime_data_df.groupby("State")


crime_rate_avg[["Violent Crime Rate", "Property Crime Rate"]].mean()
```

```python
#create bar chart
N = 10
violentcrime = (291, 330, 167, 392, 258, 635, 460, 628, 260, 117)
propertycrime = (2368, 3454, 2139, 2344, 2564, 3416, 2378, 3416 , 2804, 2460)
ind = np.arange(N)  
width = 0.35       


p1 = plt.bar(ind, propertycrime, width)
p2 = plt.bar(ind, violentcrime, width)


plt.ylabel('Total Crime Rate/ 100,000 Citizens')
plt.title('Crime Rate of States With Highest % of Drug Deaths 1999-2014')
plt.xticks(ind, ('WV', 'OH', 'NH', 'PA', 'KY', 'MD', 'MA', 'DE', 'RI', 'ME'))
plt.yticks(np.arange(0, 4000, 500))
plt.legend((p1[0], p2[0]), ('Property Crime Rate', 'Violent Crime Rate'))
plt.savefig("HighCrime.png")
plt.show()
```

```python
N = 10
violentcrime = (300, 231, 172, 499, 278, 262, 378, 303, 260, 513)
propertycrime = (3225, 1949, 2088, 4036, 2641, 2899, 3542, 3269 , 4178, 3004)
ind = np.arange(N)  
width = 0.35       


p1 = plt.bar(ind, propertycrime, width)
p2 = plt.bar(ind, violentcrime, width)


plt.ylabel('Total Crime Rate/ 100,000')
plt.title('Crime Rate for States With Lowest % of Drug Deaths 1999-2014')
plt.xticks(ind, ('NE', 'SD', 'ND', 'TX', 'IA', 'MN', 'KS', 'MS', 'HI', 'CA'))
plt.yticks(np.arange(0, 4000, 500))
plt.legend((p1[0], p2[0]), ('Property Crime Rate', 'Violent Crime Rate'))
plt.savefig("LowCrime.png")
plt.show()
```
