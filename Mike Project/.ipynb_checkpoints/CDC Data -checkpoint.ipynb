```python
import pandas as pd 
import matplotlib.pyplot as plt
import csv
```

```python
overdose_data = pd.read_csv("overdose_data_1999-2015.csv")
```

```python
male_df = overdose_data.loc[overdose_data["Overall"] == "  Male"]
female_df = overdose_data.loc[overdose_data["Overall"] == "  Female"]
drugs_df=overdose_data.loc[(overdose_data["Overall"] != "  Male") & (overdose_data["Overall"] != "  Female")]

```

```python
#x_ticks = ("1999", "2000", "2001","2002","2003","2004","2005","2006","2007","2008","2009","2010","2011","2012","2013","2014","2015")
#tick_locations = [value for value in x_ticks]
#plt.xticks(x_ticks)
labels = ("Total", "Male", "Female")
plt.figure(figsize=(10,10)) 

plt.plot(drugs_df.iloc[0,1:])
plt.plot(male_df.iloc[0,1:])
plt.plot(female_df.iloc[0,1:])
plt.legend(labels, loc = "best")
plt.ylabel("Total Overdose Deaths")
plt.xlabel("Years")
plt.show()
```
