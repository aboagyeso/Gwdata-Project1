```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
us_file="us_postal_codes.csv"
us_data=pd.read_csv(us_file)
us_data=us_data.rename(columns={"Zip Code":"Zipcode"})
us_data.dtypes
```

```python
education_data_2015= "Census_2015/census_education_2015.csv"
census_edu_2015= pd.read_csv(education_data_2015)
census_edu_2015.dtypes
```

```python
census_edu_2015.count()
```

```python
combined_edu_2015=pd.merge(us_data, census_edu_2015, on="Zipcode")
combined_edu_2015.head()
combined_edu_2015.to_csv("Census_2015/combined_edu_2015.csv")
```

```python
combined_edu_2015.count()
```

```python
grouped_edu_2015= combined_edu_2015.groupby(["County"])
grouped_edu_2015.count()
```

```python
population_sum=grouped_edu_2015["Population"].sum().tolist()
none_sum=grouped_edu_2015["None"].sum().tolist()
ged_sum=grouped_edu_2015["GED"].sum().tolist()
high_school_sum=grouped_edu_2015["High School Degree"].sum().tolist()
associate_sum=grouped_edu_2015["Associate Degree"].sum().tolist()
bachelor_sum=grouped_edu_2015["Bachelor Degree"].sum().tolist()
masters_sum=grouped_edu_2015["Masters Degree"].sum().tolist()
doctorate_sum=grouped_edu_2015["Doctorate Degree"].sum().tolist()
professional_sum=grouped_edu_2015["Professional Degree"].sum().tolist()
```

```python
county_edu_list=[]
state_list=[]
for x, group in grouped_edu_2015:
    county_edu_list.append(x)
        
    
print(len(county_edu_list))

  
```

```python
edu_2015_df=pd.DataFrame(county_edu_list, columns=["County"])
edu_2015_df["Population"]= population_sum
edu_2015_df["None"]= none_sum
edu_2015_df["GED"]= ged_sum
edu_2015_df["High School Degree"]= high_school_sum
edu_2015_df["Associate Degree"]= associate_sum
edu_2015_df["Bachelor Degree"]= bachelor_sum
edu_2015_df["Masters Degree"]= masters_sum
edu_2015_df["Professional Degree"]= professional_sum
edu_2015_df["Doctorate Degree"]= doctorate_sum

edu_2015_df
```

```python
edu_2015_df.to_csv("Census_2015/edu_2015_by_County.csv")
```
