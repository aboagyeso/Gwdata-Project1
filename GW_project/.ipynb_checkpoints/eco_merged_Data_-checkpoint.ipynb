```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
us_file="us_postal_codes.csv"
us_data=pd.read_csv(us_file)
us_data=us_data.rename(columns={"Zip Code":"Zipcode"})
us_data.dtypes
```

```python
economic_data_2015= "Census_2015/census_economic_2015.csv"
census_eco_2015= pd.read_csv(economic_data_2015)
census_eco_2015.dtypes
```

```python
census_eco_2015.count()
```

```python
combined_eco_2015=pd.merge(us_data, census_eco_2015, on="Zipcode")
combined_eco_2015.head()
combined_eco_2015.to_csv("Census_2015/combined_eco_2015.csv")
```

```python
combined_eco_2015.count()
```

```python
grouped_eco_2015= combined_eco_2015.groupby(["County"])
grouped_eco_2015.count()
```

```python
population_sum=grouped_eco_2015["Population"].sum().tolist()
household_income_sum=grouped_eco_2015["Household Income"].sum().tolist()
per_capita_sum=grouped_eco_2015["Per Capita Income"].sum().tolist()
poverty_sum=grouped_eco_2015["Poverty Count"].sum().tolist()
employable_count=grouped_eco_2015["Employable Count"].sum().tolist()
unemployable_count=grouped_eco_2015["Unemployment Count"].sum().tolist()
```

```python
county_eco_list=[]
for x, group in grouped_eco_2015:
    county_eco_list.append(x)
    
print(len(county_eco_list))
    
```

```python
eco_2015_df=pd.DataFrame(county_eco_list, columns=["County"])
eco_2015_df["Population"]= population_sum
eco_2015_df["Household Income"]= household_income_sum
eco_2015_df["Per Capita Income"]= per_capita_sum
eco_2015_df["Poverty Count"]= poverty_sum
eco_2015_df["Poverty Rate"] = 100 * \
    eco_2015_df["Poverty Count"].astype(
        int) / eco_2015_df["Population"].astype(int)

eco_2015_df["Employable Count"]=employable_count
eco_2015_df["Unemployment Count"]=unemployable_count
eco_2015_df["Unemployment Rate"] = 100 * \
    eco_2015_df["Unemployment Count"].astype(
        int) / eco_2015_df["Employable Count"].astype(int)
```

```python
eco_2015_df
```

```python
eco_2015_df.to_csv("Census_2015/eco_2015_by_County.csv")
```
